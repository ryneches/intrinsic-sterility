```python
from IPython.display import display, Image
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

figure(figsize=(16, 4))


for n,(name,scale,margin,fname) in enumerate(zip( ['(a) UV Treated', '(b) No UV', '(c) Intermediate UV', '(d) No part'], 
                                 [r'200$\mu$l', r'200$\mu$l', r'200$\mu$l', r'100$\mu$l'],
                                 [4.5,3.5,8,4],
                                 ['Fig3a.png', 'Fig3b.png', 'Fig3c.png', 'Fig3d.png' ])) : 
    i =  mpimg.imread( fname )
    subplot(1,4,n+1)
    plt.imshow(i)
    text( i.shape[1]/margin, i.shape[0]*1.15, name, fontsize=20 )
    text( 700, 520, scale, fontsize=9 )
    plt.axis('off')
        
plt.tight_layout()

savefig( 'Fig3.pdf', format='pdf', dpi=200 )
```
